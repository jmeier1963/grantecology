# Applying selenium headless chrome webdriver to scraping of grants

## Scraping of Hewlett grants
Note that the scraper was interrupted and thus two dataframes needed to be concatenated

```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
options.add_argument('headless')

# set the window size
options.add_argument('window-size=1200x600')

links = []
grantees = []
purposes = []
amounts = []
programs = []
dates = []
overviews = []

# scrape grant page and store data in lists for later aggregation into one dataframe
def parse_grant(url):
    driver = webdriver.Chrome(chrome_options=options)
    driver.get(url)
    driver.implicitly_wait(1)
    links.append(url)
    try:
        grantee = driver.find_elements_by_css_selector('h1')[0].get_attribute('innerHTML')
    except IndexError:
        grantee = " "
    grantees.append(grantee)
    
    try:
        purpose = driver.find_elements_by_css_selector('h3.large-subtitle')[0].get_attribute('innerHTML').strip()
    except IndexError:
        purpose = " "
    purposes.append(purpose)
    
    try:
        amount = driver.find_elements_by_css_selector('div.highlights-value')[0].get_attribute('innerHTML').strip()
    except IndexError:
        amount = " "
    amounts.append(amount)
    
    try:
        program_html = driver.find_elements_by_css_selector('div.highlights-value')[1].get_attribute('innerHTML')
        program = BeautifulSoup(program_html, "lxml").get_text().strip()
    except IndexError:
        program = " "
    programs.append(program)
    
    try:
        date = driver.find_elements_by_css_selector('div.highlights-value')[2].get_attribute('innerHTML').strip()
    except:
        date = " "
    dates.append(date)
    
    try: 
        description1 = driver.find_elements_by_css_selector('div.grant-overview')[0].get_attribute('innerHTML').strip()
    except IndexError:
        description1 = " "
    try:
        strategy_html = driver.find_elements_by_css_selector('a.highlight-strategy-link')[0].get_attribute('innerHTML')
        strategy = BeautifulSoup(strategy_html, "lxml").get_text().strip()
    except IndexError:
        strategy = " "
    overviews.append(description1 + " " + strategy)
    

# Get posts for each grant with links to detailed information on grant-page
def parse(url):
    driver = webdriver.Chrome(chrome_options=options)
    driver.get(url)
    # wait up to 1 secondsfor the elements to become available
    driver.implicitly_wait(3)
    posts = driver.find_elements_by_css_selector('article')
    for post in posts:
        try:
            grantee_link = post.find_elements_by_css_selector('a.listing-highlight-link')[0].get_attribute('href')
            # byline = post.find_elements_by_css_selector('div.byline-item')[0].get_attribute('innerHTML').strip()
            print grantee_link
            parse_grant(grantee_link)
        except IndexError:
            pass
    try:
        next_page = driver.find_elements_by_css_selector('div.pagination a.next')[0].get_attribute('href')
    except IndexError:
        return(None)
    print len(posts), "grants scraped from page", str(int(url[72:]))
    next_url = url[0:72] + str(int(url[72:])+1)
    return(next_url)


# Start URL and loop over next button
url = 'https://hewlett.org/grants/?sort=date&grant_programs=21943&current_page=22'

while url is not None:
    url = parse(url)

df = pd.DataFrame()
df["links"]=links
df["Grantee"]=grantees
df["Purpose"]=purposes
df["Amount"]=amounts
df["Program"]=programs
df["Date"]=dates
df["Overview"]=overviews

df.to_excel("hewlett_all_environment.xlsx")
df.info()
```

## Scraping of CIFF Environment grants

```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
options.add_argument('headless')

# set the window size
options.add_argument('window-size=1200x600')

links = []
grantees = []
purposes = []
amounts = []
programs = []
dates = []
overviews = []

# scrape grant page and store data in lists for later aggregation into one dataframe
def parse_grant(url):
    driver = webdriver.Chrome(chrome_options=options)
    driver.get(url)
    driver.implicitly_wait(1)
    links.append(url)
    try:
        grantee = driver.find_elements_by_css_selector('h1.selectionShareable')[0].get_attribute('innerHTML')
    except IndexError:
        grantee = " "
    grantees.append(grantee)
    print "Grantee", grantee
    
    try:
        purpose_html = driver.find_elements_by_css_selector('p.bold-intro.selectionShareable')[0].get_attribute('innerHTML')
        purpose = BeautifulSoup(purpose_html, "lxml").get_text().strip()
    except IndexError:
        purpose = " "
    purposes.append(purpose)
    print "Purpose", purpose
    
    try:
        amount = driver.find_elements_by_css_selector('h2.selectionShareable span')[0].get_attribute('innerHTML').strip()
    except IndexError:
        amount = " "
    amounts.append(amount)
    print "Amount", amount
    
    program = "Climate Change"
    programs.append(program)
    
    try:
        date = driver.find_elements_by_css_selector('div.intro p.selectionShareable')[0].get_attribute('innerHTML').strip()
    except:
        date = " "
    dates.append(date)
    print "Date", date
    
    try: 
        descriptions = driver.find_elements_by_css_selector('div.rich-text p.selectionShareable')
        description = " "
        for t in descriptions:
            description = description + BeautifulSoup(t.get_attribute('innerHTML'), "lxml").get_text().strip()
    except IndexError:
        description = " "
    overviews.append(description)
    print "Description", description

# Get posts for each grant with links to detailed information on grant-page
def parse(url):
    driver = webdriver.Chrome(chrome_options=options)
    driver.get(url)
    # wait up to 1 secondsfor the elements to become available
    driver.implicitly_wait(3)
    posts = driver.find_elements_by_css_selector('.td-url')
    for post in posts:
        try:
            grantee_link = post.find_elements_by_css_selector('a.selectionShareable')[0].get_attribute('href')
            print grantee_link
            parse_grant(grantee_link)
        except IndexError:
            pass
    print len(posts), "grants scraped from page", url


# Start URL and loop over next button
print "Scraping open grants"
url_open_grants = 'https://ciff.org/grant-portfolio/list/?region=&programme=climate-change&order_by=-grant_value'
parse(url_open_grants)

print "Scraping closed grants"
url_closed_grants = 'https://ciff.org/grant-portfolio/closed/?region=&programme=climate-change&order_by=-grant_value'
parse(url_closed_grants)

df = pd.DataFrame()
df["links"]=links
df["Grantee"]=grantees
df["Purpose"]=purposes
df["Amount"]=amounts
df["Program"]=programs
df["Date"]=dates
df["Overview"]=overviews

df.to_excel("ciff_all_climate.xlsx")
df.info()
```

## Scraping of Packard conservation and science grants

```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1200x600')
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

links = []
grantees = []
purposes = []
amounts = []
programs = []
dates = []
overviews = []

# scrape grant page and store data in lists for later aggregation into one dataframe
def parse_grant(url):
    driver = webdriver.Chrome(chrome_options=options)
    driver.get(url)
    driver.implicitly_wait(1)
    links.append(url)
    try:
        grantee = driver.find_elements_by_css_selector('article h2')[0].get_attribute('innerHTML')
    except IndexError:
        grantee = " "
    grantees.append(grantee)
    print "Grantee", grantee
    
    try:
        purpose = driver.find_elements_by_css_selector('article p')[1].get_attribute('innerHTML')[26:]
    except IndexError:
        purpose = " "
    purposes.append(purpose)
    print "Purpose", purpose
    
    try:
        amount = driver.find_elements_by_css_selector('article p')[2].get_attribute('innerHTML')[24:]
    except IndexError:
        amount = " "
    amounts.append(amount)
    print "Amount", amount
    
    try:
        program = driver.find_elements_by_css_selector('article p')[4].get_attribute('innerHTML')[30:]
    except IndexError:
        program = " "
    programs.append(program)
    print "Program", program
    
    try:
        date = driver.find_elements_by_css_selector('article p')[0].get_attribute('innerHTML')[14:18]
    except:
        date = " "
    dates.append(date)
    print "Date", date
    
    overviews.append(purpose)

    
# Get posts for each grant with links to detailed information on grant-page
def parse(url):
    print url
    driver = webdriver.Chrome(chrome_options=options)
    driver.get(url)
    # wait up to 1 secondsfor the elements to become available
    driver.implicitly_wait(3)
    posts = driver.find_elements_by_css_selector('tr')
    for post in posts:
        try:
            grantee_link = post.find_elements_by_css_selector('a')[0].get_attribute('href')
            print grantee_link
            parse_grant(grantee_link)
        except IndexError:
            pass
    try:
        next_url = driver.find_elements_by_css_selector('a.next.page-numbers')[0].get_attribute('href')
    except IndexError:
        return(None)
    print len(posts), "grants scraped from page", url[68:71]
    return(next_url)


# Start URL and loop over next button
url = 'https://www.packard.org/grants-and-investments/grants-database/page/1/?grant_keyword&program_area=Conservation%20and%20Science&award_amount=All%20Amounts&award_year=All%20Years'
while url is not None:
    url = parse(url)

df = pd.DataFrame()
df["links"]=links
df["Grantee"]=grantees
df["Purpose"]=purposes
df["Amount"]=amounts
df["Program"]=programs
df["Date"]=dates
df["Overview"]=overviews

df.to_excel("packard_all_environment.xlsx")
df.info()
```
